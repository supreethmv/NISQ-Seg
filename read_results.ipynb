```python
def load_gurobi_data_for_seed(seed):
    filepath = f'gurobi_{seed}.csv'
    gurobi_df = pd.read_csv(filepath)
    return gurobi_df


def get_exact_min_cut_value(row):
    gurobi_df = load_gurobi_data_for_seed(row['Seed'])
    # Find the row in the Gurobi DataFrame that matches the "No. of Pixels"
    matching_row = gurobi_df[gurobi_df['No. of Pixels'] == row['No. of Pixels']]
    if not matching_row.empty:
        # Assuming there's only one matching row per "No. of Pixels" per seed
        return matching_row.iloc[0]['Min-Cut Cost']
    return None
```

```python
abse_data['Exact Min-Cut Value'] = abse_data.apply(get_exact_min_cut_value, axis=1)
abse_data
```

```python
abse_data['Relative Error'] = (abse_data['Min-Cut Cost'] - abse_data['Exact Min-Cut Value']).abs() / abse_data['Exact Min-Cut Value'].abs()
abse_data
```

```python
# pixel_size = 16
# optimizer = 'COBYLA'

# pixel_sizes = [4,16,64]
pixel_sizes = [4,16]
# optimizers = ["COBYLA", "Powell", "CG", "BFGS", "L-BFGS-B", "SLSQP"]
optimizers = ["SLSQP"]


for pixel_size in pixel_sizes:
    for optimizer in optimizers:
        abse_errors = []
        abse_iterations = []
        acabse_errors = []
        acabse_iterations = []

        # Define your seeds
        seeds = [111, 222, 333, 444, 555]

        # Loop through the seeds and call your function
        for seed in seeds:
            abse_errors.append(np.array(list(abse_data[(abse_data['No. of Pixels'] == pixel_size) & (abse_data['Seed'] == str(seed)) & (abse_data['Optimizer Method'] == optimizer)]["Relative Error"])))
            acabse_errors.append(np.array(list(acabse_data[(acabse_data['No. of Pixels'] == pixel_size) & (acabse_data['Seed'] == str(seed)) & (acabse_data['Optimizer Method'] == optimizer)]["Relative Error"])))
            
            abse_iterations.append(np.array(list(abse_data[(abse_data['No. of Pixels'] == pixel_size) & (abse_data['Seed'] == str(seed)) & (abse_data['Optimizer Method'] == optimizer)]["Iterations"])))
            acabse_iterations.append(np.array(list(acabse_data[(acabse_data['No. of Pixels'] == pixel_size) & (acabse_data['Seed'] == str(seed)) & (acabse_data['Optimizer Method'] == optimizer)]["Iterations"])))


        # Convert lists to numpy arrays for easier calculations
        abse_errors = np.array(abse_errors)
        acabse_errors = np.array(acabse_errors)

        # Calculate statistics
        mean_abse = np.mean(abse_errors, axis=0)
        min_abse = np.min(abse_errors, axis=0)
        max_abse = np.max(abse_errors, axis=0)
        std_abse = np.std(abse_errors, axis=0)

        upper_abse = mean_abse + std_abse
        lower_abse = mean_abse - std_abse


        mean_acabse = np.mean(acabse_errors, axis=0)
        min_acabse = np.min(acabse_errors, axis=0)
        max_acabse = np.max(acabse_errors, axis=0)
        std_acabse = np.std(acabse_errors, axis=0)

        upper_acabse = mean_acabse + std_acabse
        lower_acabse = mean_acabse - std_acabse

        plt.figure(figsize=(18, 5))

        # Subplot 1: Relative Error
        plt.subplot(1, 2, 1)
        plt.plot(ns, mean_acabse, 'g-', label='ACABSE Average')
        plt.fill_between(ns, lower_acabse, upper_acabse, color='g', alpha=0.3)
        plt.plot(ns, mean_abse, 'r-', label='ABSE Average')
        plt.fill_between(ns, lower_abse, upper_abse, color='r', alpha=0.3)
        plt.title(f'Relative Error ({optimizer}, Pixel Size {pixel_size})')
        plt.xlabel('Number of Layers')
        plt.ylabel('Relative Error')
        plt.xticks(ticks=ns)
        plt.legend()




        # Convert lists to numpy arrays for easier calculations
        abse_iterations = np.array(abse_iterations)
        acabse_iterations = np.array(acabse_iterations)

        # Calculate statistics
        mean_abse_iter = np.mean(abse_iterations, axis=0)
        min_abse_iter = np.min(abse_iterations, axis=0)
        max_abse_iter = np.max(abse_iterations, axis=0)
        std_abse_iter = np.std(abse_iterations, axis=0)

        upper_abse_iter = mean_abse_iter + std_abse_iter
        lower_abse_iter = mean_abse_iter - std_abse_iter


        mean_acabse_iter = np.mean(acabse_iterations, axis=0)
        min_acabse_iter = np.min(acabse_iterations, axis=0)
        max_acabse_iter = np.max(acabse_iterations, axis=0)
        std_acabse_iter = np.std(acabse_iterations, axis=0)

        upper_acabse_iter = mean_acabse_iter + std_acabse_iter
        lower_acabse_iter = mean_acabse_iter - std_acabse_iter

        # Subplot 2: Iterations
        plt.subplot(1, 2, 2)
        plt.plot(ns, mean_acabse_iter, 'g-', label='ACABSE Average Iterations')
        plt.fill_between(ns, mean_acabse_iter - std_acabse_iter, mean_acabse_iter + std_acabse_iter, color='g', alpha=0.3)
        plt.plot(ns, mean_abse_iter, 'r-', label='ABSE Average Iterations')
        plt.fill_between(ns, mean_abse_iter - std_abse_iter, mean_abse_iter + std_abse_iter, color='r', alpha=0.3)
        plt.title(f'Iterations ({optimizer}, Pixel Size {pixel_size})')
        plt.xlabel('Number of Layers')
        plt.ylabel('Iterations')
        plt.xticks(ticks=ns)
        plt.legend()

        # Saving and showing the plot
        if not os.path.exists("plots"):
            os.makedirs("plots")
        save_file_path = os.path.join("plots", f"Comparison_{pixel_size}_{optimizer}.png")
        plt.savefig(save_file_path, dpi=300, bbox_inches='tight')
        plt.show()
```
